```python
import numpy as np
import matplotlib.pyplot as plt
```

# NOT Gate

```python
def sign(v):
	return int(v >= 0)
```

```python
def perceptron(x, w, b):
    """ Function implemented by a perceptron with 
		weight vector w and bias b """
    v = np.dot(w, x) + b
    y = sign(v)
    return y
```

```python
def NOT_Gate(x):
	return perceptron(x, w=-1, b=0.5)
```

```python
print("NOT(0) = {}".format(NOT_Gate(0)))
print("NOT(1) = {}".format(NOT_Gate(1)))
```

# AND Gate

```python
area = 200
fig = plt.figure(figsize=(6, 6))
plt.title('The AND Gate', fontsize=20)
ax = fig.add_subplot(111)
# color red: is class 0 and color blue is class 1.
ax.scatter(0, 0, s=area, c='r', label="Class 0")
ax.scatter(0, 1, s=area, c='r', label="Class 0")
ax.scatter(1, 0, s=area, c='r', label="Class 0")
ax.scatter(1, 1, s=area, c='b', label="Class 1")
plt.grid()
plt.show()
```

```python
def AND_Gate(x):
    w = np.array([1, 1])
    b = -1.5
    return perceptron(x, w, b)
```

```python
print("AND({}, {}) = {}".format(1, 1, AND_Gate(np.array([1, 1]))))
print("AND({}, {}) = {}".format(1, 0, AND_Gate(np.array([1, 0]))))
print("AND({}, {}) = {}".format(0, 1, AND_Gate(np.array([0, 1]))))
print("AND({}, {}) = {}".format(0, 0, AND_Gate(np.array([0, 0]))))
```

# OR Gate

```python
area = 200
fig = plt.figure(figsize=(6, 6))
plt.title('The OR Gate', fontsize=20)
ax = fig.add_subplot(111)
# color red: is class 0 and color blue is class 1.
ax.scatter(0, 0, s=area, c='r', label="Class 0")
ax.scatter(0, 1, s=area, c='b', label="Class 1")
ax.scatter(1, 0, s=area, c='b', label="Class 1")
ax.scatter(1, 1, s=area, c='b', label="Class 1")
plt.grid()
plt.show()
```

```python
def OR_Gate(x):
    w = np.array([1, 1])
    b = -0.5
    return perceptron(x, w, b)
```

```python
print("OR({}, {}) = {}".format(1, 1, OR_Gate(np.array([1, 1]))))
print("OR({}, {}) = {}".format(1, 0, OR_Gate(np.array([1, 0]))))
print("OR({}, {}) = {}".format(0, 1, OR_Gate(np.array([0, 1]))))
print("OR({}, {}) = {}".format(0, 0, OR_Gate(np.array([0, 0]))))
```

```python
def XOR_net(x):
    gate_1 = AND_Gate(x)
    gate_2 = NOT_Gate(gate_1)
    gate_3 = OR_Gate(x)
    new_x = np.array([gate_2, gate_3])
    output = AND_Gate(new_x)
    return output
```

```python
print("XOR({}, {}) = {}".format(1, 1, XOR_net(np.array([1, 1]))))
print("XOR({}, {}) = {}".format(1, 0, XOR_net(np.array([1, 0]))))
print("XOR({}, {}) = {}".format(0, 1, XOR_net(np.array([0, 1]))))
print("XOR({}, {}) = {}".format(0, 0, XOR_net(np.array([0, 0]))))
```
